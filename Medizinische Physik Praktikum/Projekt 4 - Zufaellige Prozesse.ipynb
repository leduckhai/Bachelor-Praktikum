Frage: Wenn n gleich 10000 ist, wie viele Kernzerf¨alle erwarten sie durchschnittlich
w¨ahrend des allerersten Zeitintervalls dieser Simulation? Erkl¨aren sie in einer
Markdownzelle.
    
    Mit dieser Formel kann man die Anzahl der Zerfälle im ersten Zeitinterval bestimmen:
        N(t)=N(0)*e**((-Lambda)*t) --> mit t=1; N(0)=10000; Lambda=0,001
        N(1)=10000*e**(-0.001)
    
    Die Anzal der Zerfall ist die Differenz zwischen N(0) und N(1)
        dN=10

```python
##Kernzerfall-Histogramm

n=10000
rep=100
decay=[0]*rep #-Liste der Werte der Zerfälle

for t in range (rep):
    for i in range (n):
        chance=int(random()*1000) #-Die Wahrscheinlichkeit für ein Kern um zu zerfallen ist 0,001
        if chance==1:
            decay[t]=decay[t]+1
        
#-Liste für Histogramm
decayList=list(range(max(decay)+1))
decayHist=[0]*(max(decay)+1) 


#-Erstellung des Histogramms
for i in range (max(decay)+1):
    for j in range (rep):
        if decay[j]==i:
            decayHist[i]=decayHist[i]+1
        else:
            decayHist[i]=decayHist[i]
            
#-Poisson
lamb=0.001*n
prob=[0]*(max(decay)+1)

def poisson():
    prob[i] = lamb**decayList[i]/factorial(decayList[i])*exp(-lamb)*rep

for i in range (max(decay)+1): 
    poisson()


plt.figure(figsize=(15,15)) 

#-Histogramm
plt.bar(decayList, decayHist, label="Monte Carlo results")
plt.title("Verteilung aller möglichen Werte der gesamten Zerfälle")
plt.xlabel("Anzahl der Zerfall")
plt.ylabel("Häufigkeit")

#-Histogramm-Poisson
plt.plot(decayList, prob, color="red")

plt.legend()
plt.show()


#-Test Liste
#print (decay)
#print (max(decay))
#print (decayList)
#print (decayHist)
#print (prob)
```
